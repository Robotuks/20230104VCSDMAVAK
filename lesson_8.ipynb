## Merge, join, concatenate

```python
df_1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
    }
)
df_1
```

```python
df_2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
    }
)
df_2
```

```python
pd.concat( [df_1, df_2], ignore_index=True )
```

```python
pd.concat( [df_1, df_2],  axis=1,  )
```

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K5"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)

right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K4"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
```

```python
left = left.set_index('key')
left
```

```python
right = right.set_index('key')
right
```

```python
left.join(right)
```

```python
left.join(right, how='outer')
```

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K5"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)

right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K4"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
left
```

```python
pd.merge(left, right, how='left', on='key')
```

```python
right = right.rename(columns={'key': 'key_1'})
right
```

```python

pd.merge(left, right, how='left', left_on='key', right_on='key_1')
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0,100)
y = x*5
z = x ** 2
```

```python
fig = plt.figure()
ax = fig.add_axes( (0,0,1,1) )

ax.plot( x, y )
ax.set_xlabel('X asis')
ax.set_ylabel('Y asis')
ax.set_title('Pavadinimas')

ax.set_xticklabels()
```

```python
fig = plt.figure()
ax = fig.add_axes( (0,0,1,1) )
print(z[50])
z[49] = 2500
z[50] = 2200

ax.plot( x, z )
ax_2 = fig.add_axes( (0.3,0.5,0.2,0.2) ) # x_start, y_start, width, height
ax_2.plot( x, z )
ax_2.set_xlim(46,52)
ax_2.set_ylim(2100,2800)
ax_2.spines[['right', 'top']].set_visible(False)
```

```python
fig, axes = plt.subplots( 3,3 )

for row in axes:
    row[2].plot(x,z)
    for col in row:
        col.plot(x,y)
```

```python
fig, axes = plt.subplots( 1,2 )

axes[0].plot(x,y, c='b', linewidth=4, linestyle='--')
axes[1].plot(x,z, c='r', linewidth=4, linestyle='-')
```
