3. ParaÅ¡ykite funkcijÄ…, kuri priims string tipo kintamÄ…jÄ¯, graÅ¾inkite kiek tame stringÄ™ yra raidÅ¾iÅ³, skaiÄiÅ³ ir skyrybos (specialiÅ³) Å¾enklÅ³, tarpÅ³.

```python
def symbol_extract(text: str):
    numbers = 0
    letters = 0
    spaces = 0
    specials = 0
    for symbol in text:
        if symbol.isalpha():
            letters += 1
        elif symbol.isdigit():
            numbers += 1
        elif symbol.isspace():
            spaces += 1
        else:
            specials +=1
    print (f'letters: {letters}, numbers: {numbers}, spaces: {spaces}, specials: {specials}')

symbol_extract("hello 148   !$%^#@^ðŸ˜ŠðŸ˜ŠðŸ˜Š")
```
